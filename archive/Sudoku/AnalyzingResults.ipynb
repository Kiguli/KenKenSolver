# Sudoku Solver Results Analysis

This notebook analyzes the performance of the neuro-symbolic Sudoku solver on 4x4 and 9x9 puzzles.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Load Results

```python
fp = "./results/"
ns_results = pd.read_csv(fp + 'neurosymbolic_solver.csv')
ns_results = ns_results.set_index('size')
ns_results = ns_results.rename(columns={'accuracy': 'accuracy (%)', 'avg_time': 'avg_time (s)'})

# Convert accuracy to percentage
ns_results['accuracy (%)'] = ns_results['accuracy (%)'] * 100

print("Neuro-Symbolic Solver Results:")
ns_results
```

## Summary Statistics

```python
print("=" * 50)
print("SUDOKU SOLVER PERFORMANCE SUMMARY")
print("=" * 50)
print(f"\nPuzzles tested per size: 100")
print(f"\nResults by grid size:")
print("-" * 50)
for size in ns_results.index:
    acc = ns_results.loc[size, 'accuracy (%)']
    time = ns_results.loc[size, 'avg_time (s)']
    print(f"  {size}x{size}: {acc:.1f}% accuracy, {time*1000:.2f}ms avg solve time")
print("-" * 50)
print(f"\nOverall accuracy: {ns_results['accuracy (%)'].mean():.1f}%")
```

## Accuracy by Grid Size

```python
x = ns_results.index

plt.figure(figsize=(8, 5))
plt.bar(x.astype(str), ns_results['accuracy (%)'], color='steelblue', edgecolor='black')
plt.xlabel('Grid Size')
plt.ylabel('Accuracy (%)')
plt.title('Sudoku Solver Accuracy by Grid Size')
plt.ylim(0, 105)
plt.grid(axis='y', alpha=0.3)

# Add value labels on bars
for i, (size, acc) in enumerate(zip(x, ns_results['accuracy (%)'])):
    plt.text(i, acc + 2, f'{acc:.0f}%', ha='center', fontweight='bold')

plt.tight_layout()
plt.show()
```

## Solving Time by Grid Size

```python
plt.figure(figsize=(8, 5))

# Convert to milliseconds for better readability
times_ms = ns_results['avg_time (s)'] * 1000

plt.bar(x.astype(str), times_ms, color='coral', edgecolor='black')
plt.xlabel('Grid Size')
plt.ylabel('Average Solve Time (ms)')
plt.title('Sudoku Solver Time by Grid Size')
plt.grid(axis='y', alpha=0.3)

# Add value labels on bars
for i, (size, time) in enumerate(zip(x, times_ms)):
    plt.text(i, time + max(times_ms)*0.02, f'{time:.1f}ms', ha='center', fontweight='bold')

plt.tight_layout()
plt.show()
```

## Comparison: 4x4 vs 9x9 Complexity

```python
time_4x4 = ns_results.loc[4, 'avg_time (s)'] * 1000
time_9x9 = ns_results.loc[9, 'avg_time (s)'] * 1000
speedup = time_9x9 / time_4x4

print("Complexity Analysis:")
print("=" * 50)
print(f"4x4 puzzle: 16 cells, 4 constraints per cell")
print(f"9x9 puzzle: 81 cells, 9 constraints per cell")
print(f"")
print(f"4x4 solve time: {time_4x4:.2f}ms")
print(f"9x9 solve time: {time_9x9:.2f}ms")
print(f"")
print(f"9x9 is {speedup:.1f}x slower than 4x4")
print(f"Cell count ratio: {81/16:.1f}x")
print(f"Constraint ratio: ~{(9*81)/(4*16):.1f}x")
```

## Key Findings

1. **Perfect Accuracy**: The Z3-based solver achieves 100% accuracy on both 4x4 and 9x9 puzzles
2. **Fast Solving**: Average solve times are under 150ms even for 9x9 puzzles
3. **Scalable**: The constraint-based approach scales well with puzzle complexity
